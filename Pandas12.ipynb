```python
import pandas as pd
employees = pd.read_csv('C:/Users/kaushik/PycharmProjects/PRJ_2/EMPLOYEES.csv')
departments = pd.read_csv('C:/Users/kaushik/PycharmProjects/PRJ_2/DEPARTMENTS.csv')
#set departement,employee_id combination as index , do not drop the index columns in the Frame (default).
employees.set_index(['DEPARTMENT_ID','EMPLOYEE_ID'],inplace = True, drop = False)

#get employees for department id 90
employees.loc[90]
#employees.stack()

#sort by department id
employees.sortlevel(0)
#sort by employee id
employees.sortlevel(1)

#calculate department level sum
employees['SALARY'].sum(level='DEPARTMENT_ID')[10]


#alternatively we can use groupby() with various column combinations.
employees.groupby(by=['DEPARTMENT_ID','HIRE_DATE'])['SALARY'].max()

#reset the index and use drop = True to avoid existing index columns
employees.reset_index(drop=True, inplace=True)
employees
```

```python
#inner join employees with departments.
pd.merge(employees,departments, on='DEPARTMENT_ID')

#Left outer join departments and employees
pd.merge(employees,departments, on='DEPARTMENT_ID',how='left').set_index(['DEPARTMENT_ID','DEPARTMENT_NAME','EMPLOYEE_ID'])

#Right outer join departments and employees usin how= (default is inner)
pd.merge(employees,departments, on='DEPARTMENT_ID',how='right').set_index(['DEPARTMENT_ID','DEPARTMENT_NAME','EMPLOYEE_ID'])
```

```python
employees.reset_index(inplace = True)
departments.reset_index(inplace=True)

#merge on index values.
pd.merge( employees.set_index('EMPLOYEE_ID'), departments.set_index('DEPARTMENT_ID') ,right_index=True, left_on='DEPARTMENT_ID')

employees.set_index('EMPLOYEE_ID', inplace = True, drop=False)
departments.set_index('DEPARTMENT_ID', inplace = True, drop = False)

#use join objects on instance. Since it is a self join , lsuffix, rsuffix requried for differentiating columns.
employees.join ( employees , lsuffix='_left', rsuffix='_right')
```
