```python
import pandas as pd
employees = pd.read_csv('C:/Users/kaushik/PycharmProjects/PRJ_2/EMPLOYEES.csv')
departments = pd.read_csv('C:/Users/kaushik/PycharmProjects/PRJ_2/DEPARTMENTS.csv')
```

```python
employees.set_index('EMPLOYEE_ID', inplace = True, drop = False)
departments.set_index('DEPARTMENT_ID', inplace = True, drop = False)
```

```python
#construct a series from 10 elements of salary and departments
salary_series = employees.SALARY[0:10]
department_series = employees.DEPARTMENT_ID[0:10]

#append the series using concat
pd.concat([salary_series,department_series])


#append the series using concat and axis =1 (columns) .it yields a Dataframe.Give a salary increment of 25% 
# and change the serries of departments.
pd.concat([salary_series * 1.25,department_series + 2000], axis =1)

#construct from last 10 departments.
department_series2 = employees.DEPARTMENT_ID[-1:-10:-1]
department_series2
#employees[employees['DEPARTMENT_ID'] == 206 ]

#inner join.  will yield now rows as keys don't overlap.
pd.concat([salary_series ,department_series2 ], axis =1, join='inner')

#use keys= identify the key from particular series in the concatenated series.
pd.concat([salary_series,department_series],keys= [ 'SALARY', 'DEPARTMENT_ID'] )

df1 = pd.concat([salary_series ,department_series ], axis =1)

df2 = pd.concat([salary_series ,department_series2 ], axis =1)

#similarly two dataframes can be concatenated. use keys = to differentiate the heading.

pd.concat([df1,df2], axis=1, keys = ['DEP_F_10','DEP_L_10'])
```
